## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)

plt.hist(n, bins=50, color='skyblue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
plt.scatter(df['x'], df['y'], label='Data Points', color='blue')

coefficients = np.polyfit(df['x'], df['y'], 1)
poly_line = np.poly1d(coefficients)
trendline_x = np.linspace(df['x'].min(), df['x'].max(), 50)
trendline_y = poly_line(trendline_x)

plt.plot(trendline_x, trendline_y, label=f'Trendline: y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}', color='red')

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatterplot with Trendline')
```
